```python
import pandas as pd
import re

data = pd.read_json ("data_5scheduler.json")
data
```

```python
data["source"].value_counts()
```

```python
# Remove duplicate courses, most likely multiple sections for the same course

data = data.drop_duplicates("identifier")
data
```

```python
# Courses rgar require a fee
data.loc[data["fee"] > 0]
```

```python
# Total courses offered by each college

data["source"].value_counts()
```

```python
# Total majors by which courses are categorized 

data["identifier"].value_counts()
course_names = data["identifier"].values.tolist()
course_names

course_string = ' '.join(course_names)
#print(course_string)


by_major = re.findall(" [A-Z]+[^-]", course_string)
majors = list(Counter(by_major).keys())
len(majors)
```

```python
# Total courses by major

from collections import Counter

major_freq = Counter(by_major)
major_freq
```
