```python
import numpy
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data_science/data.csv', error_bad_lines=False)
```

```python
df #imported dataset
```

```python
#drop columns that aren't important: classificationOfInstructionalProgramCode, externalId (same as courseNumber), subjectAbbreviation, dsubjectAbbreviation, same as departmentExternalId, institutionExternalId (same as facilityExternalId)
df.drop(['externalId', 'classificationOfInstructionalProgramCode', 'subjectAbbreviation', 'departmentExternalId', 'institutionExternalId'],  axis=1, inplace=True)
```

```python
#rename columns to be more helpful
df = df.rename(columns={"courseNumber": "Course Number", "courseTitle": "Course Title", "facilityExternalId": "Campus", "description" : "Description"})
```
